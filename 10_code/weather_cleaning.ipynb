```python
import requests

#store the data we get as a dataframe
import pandas as pd

#convert the response as a strcuctured json
import json

#mathematical operations on lists
import numpy as np

#parse the datetimes we get from NOAA
from datetime import datetime

#add the access token you got from NOAA
Token = 'VSKeYyTUCtoWpuWvDhJswcCGPdgxbrma'

station_id = "GHCND:USW00014819"
```

```python
def get_weather(station_id, bottom, top):
    # initialize lists to store data
    # make the api call
    r = requests.get(
        f"https://www.ncdc.noaa.gov/cdo-web/api/v2/data?datasetid=GHCND&dataypeid=TAVG&limit=1000&stationid={station_id}&startdate="
        + bottom
        + "&enddate="
        + top,
        headers={"token": Token},
        timeout=60,
    )
    # load the api response as a json
    d = json.loads(r.text)
    data = pd.DataFrame.from_dict(d["results"])
    time.sleep(1)
    return data
```

```python
station_ids = {
    "Chicago": "GHCND:USW00014819",
    "Philadelphia": "GHCND:USC00366886",
    "Boston": "GHCND:USW00014739",
    "New York City": "GHCND:USW00014732",
    "Washington, D.C.": "GHCND:USW00013743",
    "Atlanta": "GHCND:USW00013874",
    "Los Angeles": "GHCND:USW00023174",
    "San Francisco": "GHCND:USW00023272",
    "Portland": "GHCND:USW00024229",
}
month_cuts = [
    ("-01-01", "-02-28"),
    ("-03-01", "-04-30"),
    ("-05-01", "-06-30"),
    ("-07-01", "-08-31"),
    ("-09-01", "-10-31"),
    ("-11-01", "-12-31")
]
def get_segment(key, value, begin, end):
    dfs  = []
    for year in range(begin,end):
        print(f'Starting year {year} in city {key}')
        year_list = []
        for tuple in month_cuts:
            print(f'starting tuple beginning in month {tuple[0][1:3]}')
            beginning = str(year) + tuple[0]
            end = str(year) + tuple[1]
            d = get_weather(value, beginning, end)
            year_list.append(d)
            pass
        year_df = pd.concat(year_list)
        year_df['City'] = key
        dfs.append(year_df)
    return dfs
```

```python
chicago_2_8 = get_segment('Chicago', 'GHCND:USW00014819', 2002, 2008)
chicago_8_10 = get_segment('Chicago', 'GHCND:USW00014819', 2008, 2010)
chicago_10_12 = get_segment('Chicago', 'GHCND:USW00014819', 2010, 2012)
chicago_12_14 = get_segment('Chicago', 'GHCND:USW00014819', 2012, 2014)
chicago_14_16 = get_segment('Chicago', 'GHCND:USW00014819', 2014, 2016)
chicago_16_18 = get_segment('Chicago', 'GHCND:USW00014819', 2016, 2018)
chicago_18_20 = get_segment('Chicago', 'GHCND:USW00014819', 2018, 2020)
```

```python
chicagos = chicago_2_8 + chicago_8_10 + chicago_10_12 + chicago_12_14 + chicago_14_16 + chicago_16_18 + chicago_18_20
```

```python
chicago_final = pd.concat(chicagos)
```

```python
chicago_final.to_parquet("../00_source/chicago_weather.parquet", engine="fastparquet")
```
