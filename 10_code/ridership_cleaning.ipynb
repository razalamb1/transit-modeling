```python
import pandas as pd
```

Read in the unlinked passenger trips spreadsheet.

```python
trips = pd.read_excel("../00_source/september_2020.xlsx", sheet_name="UPT")
```

Now need to subset for appropriate transit agencies: Atlanta, Chicago, Boston, Philadelphia, Washington D.C., New York, Los Angeles, Portland, San Fransisco.

```python
keep = [
    "Chicago Transit Authority",
    "Los Angeles County Metropolitan Transportation Authority ",
    "Massachusetts Bay Transportation Authority",
    "Washington Metropolitan Area Transit Authority",
    "MTA New York City Transit",
    "Tri-County Metropolitan Transportation District of Oregon",
    "Metropolitan Atlanta Rapid Transit Authority",
    "San Francisco Bay Area Rapid Transit District",
    "City and County of San Francisco",
    "Southeastern Pennsylvania Transportation Authority"
]
```

```python
trips = trips[trips['Agency'].isin(keep)]
```

```python
trips = trips.drop(
    [
        "5 digit NTD ID",
        "4 digit NTD ID",
        "Active",
        "Reporter Type",
        "UZA",
        "UZA Name",
        "TOS",
    ],
    axis=1,
)
trips = trips[trips["Modes"].isin(["HR", "LR", "SR"])]
```

```python
trips.loc[
    trips.loc[:, "Agency"] == "City and County of San Francisco", "Agency"
] = "San Francisco Bay Area Rapid Transit District"

convert = keep = {
    "Chicago Transit Authority": "Chicago",
    "Los Angeles County Metropolitan Transportation Authority ": "Los Angeles",
    "Massachusetts Bay Transportation Authority": "Boston",
    "Washington Metropolitan Area Transit Authority": "Washington, D.C.",
    "MTA New York City Transit": "New York City",
    "Tri-County Metropolitan Transportation District of Oregon": "Portland",
    "Metropolitan Atlanta Rapid Transit Authority": "Atlanta",
    "San Francisco Bay Area Rapid Transit District": "San Francisco",
    "Southeastern Pennsylvania Transportation Authority": "Philadelphia"
}
trips["City"] = trips["Agency"].map(convert)
```

```python
trips
```

```python
trips = trips.groupby(["City", "Agency"]).sum().reset_index()
```

```python
trips = trips.melt(
    id_vars=["Agency", "City"], value_name="Total Trips", var_name="Date"
)
```

```python
trips["Month"] = trips["Date"].str[:3]
trips["Year"] = trips["Date"].str[3:].astype(int) + 2000
trips = trips[trips["Year"] < 2020]
trips.drop("Date", axis = 1, inplace=True)
```

```python
trips["Average Daily Ridership"] = 0
trips["Month Numeric"] = ""
months = {
    "JAN": ("January", 1, 31),
    "FEB": ("February", 2, 28),
    "MAR": ("March", 3, 31),
    "APR": ("April", 4, 30),
    "MAY": ("May", 5, 31),
    "JUN": ("June", 6, 30),
    "JUL": ("July", 7, 31),
    "AUG": ("August", 8, 31),
    "SEP": ("September", 9, 30),
    "OCT": ("October", 10, 31),
    "NOV": ("November", 11, 30),
    "DEC": ("December", 12, 31),
}
```

```python
for i in months.keys():
    answer = months[i]
    trips.loc[trips.loc[:, "Month"] == i, "Month Numeric"] = answer[1]
    trips.loc[trips.loc[:, "Month"] == i, "Average Daily Ridership"] = round(
        trips["Total Trips"] / answer[2], 0
    )
    trips.loc[trips.loc[:, "Month"] == i, "Month"] = answer[0]
    pass
trips.loc[
    (trips.loc[:, "Month Numeric"] == 2) & (trips.loc[:, "Year"] % 4 == 0),
    "Average Daily Ridership",
] = round(trips["Total Trips"] / 29, 0)
trips.drop("Total Trips", axis=1, inplace=True)
```

```python
assert ((trips["Month Numeric"] >= 1) & (trips["Month Numeric"] <= 13)).all()
assert (trips["Average Daily Ridership"] > 0).all()
assert not trips.isnull().values.any()
```

```python
trips.to_parquet("../20_intermediate_data/trips.parquet", engine="fastparquet")
```
